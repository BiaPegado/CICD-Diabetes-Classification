```python
import pandas as pd
from pycaret import *

# Carregar o dataset
df = pd.read_csv('Data/diabetes_prediction_dataset.csv')

# Informações gerais do dataset
df.info()

# Contagem dos valores únicos na coluna diabetes
class_counts = df['diabetes'].value_counts()

# Exibir a contagem de cada classe
print("\nContagem das Classes na Coluna 'diabetes':")
print(f"Classe 0 (Não diabético): {class_counts[0]}")
print(f"Classe 1 (Diabético): {class_counts[1]}")
```

```python
from sklearn.model_selection import train_test_split

# Separar classes
class_majority = df[df['diabetes'] == 0]
class_minority = df[df['diabetes'] == 1]

# Subamostrar a classe majoritária
class_majority_downsampled = resample(class_majority,
                                      replace=False,
                                      n_samples=len(class_minority),
                                      random_state=42)

# Combinar classes
df = pd.concat([class_majority_downsampled, class_minority])

X = df.drop("diabetes", axis=1).values
y = df.diabetes.values

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=125
)
df.head()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from sklearn.ensemble import AdaBoostClassifier
from sklearn.preprocessing import OneHotEncoder

# Definir as colunas categóricas e numéricas
num_col = ['age', 'bmi', 'HbA1c_level', 'blood_glucose_level']
cat_col = ['gender', 'smoking_history']

# Preprocessamento das colunas
transform = ColumnTransformer(
    [
        ("encoder", OrdinalEncoder(), cat_col),  # Codificação ordinal para colunas categóricas
        ('onehot', OneHotEncoder(handle_unknown='ignore'), cat_col),
        ("num_imputer", SimpleImputer(strategy="median"), num_col),  # Imputação para dados numéricos
        ("num_scaler", StandardScaler(), num_col)  # Normalização para dados numéricos
    ]
)

# Carregar os dados
df = pd.read_csv('Data/diabetes_prediction_dataset.csv')  # Substitua pelo caminho correto do dataset

# Separar a variável alvo (diabetes) das features
X = df.drop(columns=['diabetes'])
y = df['diabetes']

# Construir a pipeline
pipeline = Pipeline(steps=[
    ('preprocessing', transform),  # Preprocessamento
    ('model', AdaBoostClassifier(n_estimators=50, random_state=123))  # Modelo final
])

# Realizar o ajuste (fit) no pipeline
pipeline.fit(X, y)

# Realizar previsões (se necessário)
predictions = pipeline.predict(X)
```

```python
from pycaret.classification import *

# Criação do objeto de experimento
exp = ClassificationExperiment()

# Inicialização do setup no objeto
exp.setup(df, target='diabetes', session_id=123)

# Comparar modelos
best = exp.compare_models()
best_tuned = exp.tune_model(best)
print(best_tuned)
```

```python
# plot feature importance
exp.plot_model(best, plot = 'feature')
```

```python
final_model = exp.finalize_model(best_tuned)
exp.save_model(final_model, 'diabetes_model')

# Avalie o modelo em um conjunto de testes
evaluation_results = exp.evaluate_model(final_model)
```
